```python
import warnings
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
from itertools import product
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.arima_model import ARIMA
import calendar
from datetime import datetime, timedelta
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

## 1. Data load

```python
data = pd.read_csv('shanghai_index_1990_12_19_to_2020_03_12.csv', usecols=['Timestamp', 'Price'])
                                                                    # 只读取两列
data.Timestamp = pd.to_datetime(data.Timestamp)                     # 将数据转化成datetime格式
data = data.set_index('Timestamp')                                  # 设置index为Timestamp
data['Price'] = data['Price'].apply(pd.to_numeric, errors='ignore') # 将Price转化为numeric格式

data.head(5)
```

### 1.1 data pre process

```python
data.isnull().sum()
# 进行线性插补缺漏值
# data.Price.interpolate(inplace=True)
```

**查看数据是否有空值，如果有，就用线性插值进行补全。这里的数据没有空值，所以不进行补全操作。**

### 1.2 show seasonal information of data

```python
result = sm.tsa.seasonal_decompose(data.Price, freq=288)
result.plot()
plt.show()
```

- observed是原始数据走势图
- Trend是平滑之后的数据走势图
- seasonal是观测到的数据的季节性/周期性
- residual是残差图

```python
type(result)
```

### 1.3 data resample

**data resample by year**

```python
data_year = data.resample('Y').mean()
data_year.head(4)
```

**data resample by quarter**

```python
data_quarter = data.resample('Q').mean()
data_quarter.head(4)
```

**data resample by month**

```python
data_month = data.resample('M').mean()
data_month.head(4)
```

## 2. ARMA

### 2.1 define the parameter list

```python
warnings.filterwarnings('ignore')

ps = range(0,5)
qs = range(0,5)
parameters = list(product(ps,qs))
print(parameters)
```

### 2.2 fit the model

```python
result = []
best_aic = float("inf")
for parame in parameters:
    try:
        model = ARMA(data_month.Price,order=parame).fit()
    except ValueError:
        print("error",parame)
        continue        
    aic = model.aic
    if aic < best_aic:
        best_ARMA = model
        best_aic = aic
        best_parame = parame
    result.append([parame,aic])

print("paramer and AIC in each iteration:")
result
```

```python
print("best ARMA model:",best_ARMA.summary())
```

## 3. ARIMA

### 3.1 define the parameter list

```python
ps = range(0, 5)
ds = range(1, 3)
qs = range(0, 5)
parameters = product(ps, ds, qs)
parameters_list = list(parameters)

print(parameters_list)
```

### 3.2 fit the model

```python
result = []
best_aic = float('inf')

for param in parameters_list:
    try:
        #model = ARIMA(df_month.Price,order=(param[0], param[1], param[2])).fit()
        # SARIMAX 包含季节趋势因素的ARIMA模型
        model = sm.tsa.statespace.SARIMAX(data_month.Price,
                                order=(param[0], param[1], param[2]),
                                #seasonal_order=(4, 1, 2, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False).fit()

    except ValueError:
        print('error:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_ARIMA = model
        best_aic = aic
        best_param = param
    result.append([param, model.aic])

print("paramer and AIC in each iteration:")
result
```

```python
print('best ARIMA model:', best_ARIMA.summary())
```

## 4. Model predict

### 4.1 ARMA model

```python
model = best_ARMA
model_ARMApredict = model.predict(0,len(data_month)+12)

plt.figure(figsize=(20,6))
plt.plot(data_month, label='original')
plt.plot(model_ARMApredict, label='predict', color='r', linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

### 4.2 ARIMA model

```python
model = best_ARIMA
model_ARIMApredict = model.predict(0,len(data_month)+12)

plt.figure(figsize=(20,6))
plt.plot(data_month, label='original')
plt.plot(model_ARIMApredict, label='predict', color='r', linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

**我们可以看到，两个模型预测的结果似乎不一致，一个有向上的趋势，一个是向下的趋势。我们对这个两个结果进行可视化对比。**

### 4.3 model comparing

```python
pd.concat([model_ARMApredict.tail(12), model_ARIMApredict.tail(12)], axis=1)
```

```python
plt.plot(model_ARMApredict.tail(12), label='ARMA')
plt.plot(model_ARIMApredict.tail(12), label='ARIMA', color='r', alpha=0.5)
plt.legend()
plt.show()
```
