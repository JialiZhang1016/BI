## 2. Model training

### 2.1 Data prepare
#### 2.1.1 separate the variables and target

```python
X = df.iloc[:, :-1] 
y = df.iloc[:, -1]
# [:,:-1]ç¬¬ä¸€ä¸ªä½ç½®ä¸ºè¡Œï¼Œå–å…¨éƒ¨è¡Œï¼›
# ç¬¬äºŒä¸ªä½ç½®ä¸ºåˆ—ï¼Œä»Žç¬¬ä¸€åˆ—å–åˆ°å€’æ•°ç¬¬äºŒåˆ—
```

```python
print("show the variables: \n",X.head())
```

```python
print("show the target: \n",y.head())
```

#### 2.1.2 standerdize the variables

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
print(X)
```

#### 2.1.3 encoder the target

```python
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
print(y)
```

#### 2.1.4 separate the train and test data set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

train_data = xgb.DMatrix(X_train, label=y_train)
test_data = xgb.DMatrix(X_test, label=y_test)
```

### 2.2 Algorithm prepare

```python
param = {'boosting_type':'gbdt',
                         'objective' : 'binary:logistic', # ä»»åŠ¡ç›®æ ‡
                         'eval_metric' : 'auc',           # è¯„ä¼°æŒ‡æ ‡
                         'eta' : 0.01,                    # å­¦ä¹ çŽ‡
                         'max_depth' : 15,                # æ ‘æœ€å¤§æ·±åº¦
                         'colsample_bytree':0.8,          # è®¾ç½®åœ¨æ¯æ¬¡è¿­ä»£ä¸­ä½¿ç”¨ç‰¹å¾çš„æ¯”ä¾‹
                         'subsample': 0.9,                # æ ·æœ¬é‡‡æ ·æ¯”ä¾‹
                         'subsample_freq': 8,             # boostingçš„æ¬¡æ•°
                         'alpha': 0.6,                    # L1æ­£åˆ™
                         'lambda': 0,                     # L2æ­£åˆ™
        }
```

### 2.3 Model training

xgboost.train(params, dtrain, num_boost_round=10, evals=(), obj=None, feval=None, maximize=False, early_stopping_rounds=None, evals_result=None, verbose_eval=True, xgb_model=None, callbacks=None)

Parameters
- params (dict) â€“ Booster params.
- dtrain (DMatrix) â€“ Data to be trained.
- num_boost_round (int) â€“ Number of boosting iterations.
- evals (list of pairs (DMatrix, string)) â€“ List of validation sets for which metrics will evaluated during training. Validation metrics will help us track the performance of the model.

```python
model = xgb.train(param, 
                  train_data, 
                  evals=[(train_data, 'train'), (test_data, 'test')], 
                  num_boost_round = 5000, 
                  early_stopping_rounds=20, 
                  verbose_eval=25)
```

**The model shows, the model stop training after 25th iteration, and the final test-auc is 0.99782.  
Model training has random part, so you'd better set a seed.**
